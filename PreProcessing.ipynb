```python
#############################################  DATASET #######################
#RG
import cv2
import matplotlib.pyplot as plt

########### HSV => 0,1,2
##########  BGR =>0,1,2
#################################################   TRAIN IMAGES ########################
#### Extracting The S component from HSV
path='Images/train/'
path2='Images/trainf/'
for i in range(195):
    
    
    ###### Extracting S COMPONENT FROM HSV ########## 
    img=cv2.imread(path+str(i)+'.bmp')
    img1= cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
    #plt.imshow(img1)
    #plt.show()
    #h,s,v = cv2.split(img1)
    s_part=img1[:,:,1]
    

    #### Extracting The R component from BGR
    img2=cv2.imread(path+str(i)+'.bmp')
    R_part=img2[:,:,2]
    


    #### Extracting The G component from BGR
    img2=cv2.imread(path+str(i)+'.bmp')
    g_part=img2[:,:,1]
    

    #### MERGING S,G,R Components
    new_image=cv2.merge((s_part,R_part,g_part))
    cv2.imwrite(path2+str(i)+'.bmp',new_image)
    
 
    
```
