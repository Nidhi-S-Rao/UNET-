```python
#Find dice coefficient
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
total=0
np.set_printoptions(threshold=np.inf)
for i in range(55):
    
    
    
    
    img1=plt.imread('Images/resize/'+str(i)+'.png')
    img1=np.asarray(img1)
    #print(img1)
    img1=np.where(img1<0.9,0,1)
    #img1=img1.astype(np.bool)
    #print(img1)
    #print(img1.shape)
  
    img2=plt.imread('Images/segment_binaryf3/'+str(i)+'(o).png')
    img2=np.asarray(img2)
    img2=np.where(img2<0.9,0,1)
    #print(img2)
    #img2=img2.astype(np.bool)
    #print(img2)
    #print(img2.shape)


    intersection = np.logical_and(img1, img2)
    val= 2. * intersection.sum() / (img1.sum() + img2.sum())
    print(i,val)
    total=total+val
    
print("Dice coeff=",(total/55))
    

```

```python

#Find jaccard coefficient
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
total=0
np.set_printoptions(threshold=np.inf)
for i in range(55):
    
    
    
    
    img1=plt.imread('Images/resize/'+str(i)+'.png')
    img1=np.asarray(img1)
    #print(img1)
    img1=np.where(img1<0.6,0,1)
    #img1=img1.astype(np.bool)
    #print(img1)
    #print(img1.shape)
  
    img2=plt.imread('Images/segment_binaryf3/'+str(i)+'(o).png')
    img2=np.asarray(img2)
    img1=np.where(img1<0.6,0,1)
    #print(img2)
    #img2=img2.astype(np.bool)
    #print(img2)
    #print(img2.shape)


    intersection = np.logical_and(img1, img2)
    union=np.logical_or(img1,img2)
    val= intersection.sum() / float(union.sum())
    total=total+val
    print(val)
    
print("Jaccard coeff=",(total/55))
    

```
